```python
def rosenbrock(x0, x1):
    y = 100 * (x1 - x0 ** 2) ** 2 + (x0 - 1) ** 2
    return y
x0 = Variable(np.array(0.0))
x1 = Variable(np.array(2.0))

y=rosenbrock(x0,x1)
# 足し算、引き算、掛け算の微分が呼び出される
y.backward()
x0.grad
```

## 勾配降下法
勾配方向と逆方向に学習率×勾配値分進む
$$
 x←x-\alpha f'(x)
$$

```python
x0 = Variable(np.array(0.0))
x1 = Variable(np.array(2.0))
lr=0.001 #学習率
iters=1000 #　繰り返す回数
for i in range(iters):
    print(x0, x1)

    # 順伝播
    y = rosenbrock(x0, x1)

    # 勾配は次々と加算されていくので初期化する
    x0.cleargrad()
    x1.cleargrad()
    # 勾配計算
    y.backward()

    # 勾配方向と逆向きに進む
    x0.data -= lr * x0.grad
    x1.data -= lr * x1.grad
```

## step29 ニュートン法
２階微分を利用して、人の手によって決定するパラメータを減らしている。
$$
 x←x-\frac{f'(x)}{f''(x)} 
$$

```python

def f(x):
    y = x ** 4 - 2 * x ** 2
    return y


def gx2(x):
    return 12 * x ** 2 - 4


x = Variable(np.array(2.0))
iters = 10

for i in range(iters):
    print(i, x)

    y = f(x)
    x.cleargrad()
    y.backward()

    x.data -= x.grad / gx2(x.data)
```
